### The state space model for the discrete dynamics of the system will be exactly similar to the previous Exercise

````python
A_wiener = np.array([[0., 1.],[0., 0.]])
B_wiener = np.array([[0.],[1.]])
process_noise_variance = 0.1
measurement_noise_variance = 1.
dt_wiener  = 0.01
F_wiener = np.array([[1., dt_wiener ],[0., 1.]])
Q_wiener_discrete = process_noise_variance * np.array([[(dt_wiener **3)/3, (dt_wiener **2)/2],[(dt_wiener **2)/2,dt_wiener ]])
cov_init_wiener = 0.01*np.eye(2)
x_wiener_init = np.array([0.,1.])
t_wiener = np.arange(1001)*dt_wiener 
R_wiener = measurement_noise_variance*np.eye(1)
q_wiener = np.random.multivariate_normal(np.zeros(2),Q_wiener_discrete,(t_wiener.shape[0]))
r_wiener = np.random.multivariate_normal(np.zeros(1),R_wiener,t_wiener.shape[0])
````

````python
def propagate_linear_discrete_system_with_noise(F:np.ndarray,x_init:np.ndarray,v:np.ndarray):
    x = np.zeros((v.shape[0],x_init.shape[0]))
    x[0] = x_init
    for i in range(v.shape[0]-1):
        x[i+1] = F@x[i]+v[i]
    return x
````

````python
x_wiener = propagate_linear_discrete_system_with_noise(F_wiener,x_wiener_init,q_wiener)
````

### Define measurement model

$
y_k = h(x_k) + r_k.
$

````python
def _h_wiener(z:np.ndarray):
    return np.sin(z[0,np.newaxis])

h_wiener = np.vectorize(_h_wiener, signature='(n)->(1)') ##Vectorize the stuff to avoid doing looping

y_wiener = h_wiener(x_wiener)+r_wiener
````

### The Jacobian

$H(x_k) = \dfrac{\partial h}{\partial x_k}$

here $x_k$ is $[p(t_k) \; v(t_k)]$. Since we have $h(x_k) = \sin(p(t_k))$

$
K = P H^\top S^{-1}
$
usually, in matlab you use
```matlab
K = P H'/S
```

in `numpy` you don't have `/` operator, instead you have to use `solve`:

$
x = A^{-1}b
$

you could use `np.linalg.inv` O(n^3)

````python
def dh_dx_wiener(x):
    return np.array([[np.cos(x[0]),0.]])
````

### The prediction step is exactly the Kalman filter step since the dynamic model is linear

### Here we implement the Extended Kalman filter for general time invariant case, where the dynamic can be nonlinear

````python
def _f_wiener(x:np.ndarray,u:np.ndarray):
    return F_wiener@x

f_wiener = np.vectorize(_f_wiener,signature='(n),(m)->(n)')

#The jacobian is the F matrix
def df_dx_wiener(x:np.ndarray,u:np.ndarray):
    return F_wiener
````

````python
def Extended_Kalman_filter(f:Callable[[np.ndarray,np.ndarray],np.ndarray],h:Callable[[np.ndarray],np.ndarray],
                  df_dx:Callable[[np.ndarray,np.ndarray],np.ndarray],dh_dx:Callable[[np.ndarray],np.ndarray],
                  x_init,cov_init,Q_discrete,R_discrete,y,u=None):
    length = y.shape[0]
    x = np.zeros((length,x_init.shape[0]))
    P = np.zeros((length,x_init.shape[0],x_init.shape[0]))
    x[0] = x_init
    P[0] = cov_init
    
    if u is None:
        u = np.zeros((length,1))
        
    for i in range(length-1):
        x_     = f(x[i],u[i])
        F_x    = df_dx(x[i],u[i])
        P_     = F_x@P[i]@F_x.T + Q_discrete
        
        H_x    = dh_dx(x_)
        S      = H_x@P_@H_x.T + R_discrete
        K_tran = sla.solve(S,H_x@P_)
        P[i+1] = P_ - K_tran.T@S@K_tran
        x[i+1] = x_ + K_tran.T@(y[i]-h(x_))
    return x, P
````

````python
x_wiener_ekf_estimate,P_wiener_ekf_estimate = Extended_Kalman_filter(f_wiener,h_wiener,df_dx_wiener,dh_dx_wiener,
                                               x_wiener_init,cov_init_wiener,Q_wiener_discrete,
                                               R_wiener,y_wiener)
stdev_wiener_ekf = np.sqrt(np.diagonal(P_wiener_ekf_estimate,axis1=1,axis2=2))
````

### The unscented Kalman filter

````python
def prepare_unscented_weigths(n_state:int, alpha:float, beta:float):
    kappa = 3 - n_state
    lamb = -n_state + (n_state+kappa)*alpha**2 
    wm_0 = lamb / (n_state+ lamb)
    wp_0 = lamb / (n_state+ lamb) + (1- alpha**2 + beta)
    wm_else = 1/(2*(n_state+lamb))
    wp_else = wm_else
    WM = np.zeros(2*n_state+1)
    WP = np.zeros(2*n_state+1)
    for j in range(2*n_state + 1):
        if j==0:
            WM[j] = wm_0
            WP[j] = wp_0
        else:
            WM[j] = wm_else
            WP[j] = wp_else
    
    return lamb,WM,WP


def unscented_transformation(x:np.ndarray,P:np.ndarray,lamb:float):
    sqrt_P    = np.linalg.cholesky(P)
    return x + np.sqrt(x.shape[0]+lamb)*np.vstack([np.zeros((1,x.shape[0])),sqrt_P,-sqrt_P])

def _outer(x:np.ndarray,y:np.ndarray):
    return np.outer(x,y)
outer_prod = np.vectorize(_outer,signature="(n),(m)->(n,m)")

def compute_covariance(xs:np.ndarray,ys:np.ndarray,weight:np.ndarray):
    unnormalized = outer_prod(xs,ys)
    return np.tensordot(weight,unnormalized,axes=([0],[0]))
    
````

````python
def unscented_Kalman_filter(f:Callable[[np.ndarray],np.ndarray],h:Callable[[np.ndarray],np.ndarray],
                  WM:np.ndarray, WP:np.ndarray,lamb:float,
                  x_init,cov_init,Q_discrete,R_discrete,y,u=None):
    length = y.shape[0]
    x = np.zeros((length,x_init.shape[0]))
    P = np.zeros((length,x_init.shape[0],x_init.shape[0]))
    x[0] = x_init
    P[0] = cov_init
    if u is None:
        u = np.zeros((length,1))
    for i in range(length-1):
        x_ust = unscented_transformation(x[i],P[i],lamb)
        f_ust = f(x_ust,u[i])
        x_hat_min = WM@f_ust
        P_min = compute_covariance(f_ust-x_hat_min,f_ust-x_hat_min,WP)+Q_discrete
        
        #this is additional step mentioned in the lecture slide
        #The x_ust only used to compute h_ust
        x_ust = unscented_transformation(x_hat_min,P_min,lamb)
        x_ = WM@x_ust
        #-----------
        
        
        h_ust = h(x_ust)
        h_ = WM@h_ust
        S_ = compute_covariance(h_ust-h_,h_ust-h_,WP)+R_discrete
        cov_h_f = compute_covariance(h_ust-h_,x_ust-x_hat_min,WP)
        K_tran = sla.solve(S_,cov_h_f)
        
        #The remaining is just the ordinary kalman filter
        P[i+1] = P_min - K_tran.T@S_@K_tran
        x[i+1] = x_hat_min + K_tran.T@(y[i]-h_)        
    return x, P
````

````python
lamb,WM,WP = prepare_unscented_weigths(n_state=x_wiener_init.shape[0], alpha=1, beta=0)
x_wiener_ukf_estimate,P_wiener_ukf_estimate = unscented_Kalman_filter(f_wiener,h_wiener,WM,WP,lamb,
                                               x_wiener_init,Q_wiener_discrete,Q_wiener_discrete,
                                               R_wiener,y_wiener)
stdev_wiener_ukf = np.sqrt(np.diagonal(P_wiener_ukf_estimate,axis1=1,axis2=2))
````

````python
RMSE_UKF = np.sqrt(np.sum(np.square(x_wiener_ukf_estimate-x_wiener))/x_wiener.shape[0])
RMSE_EKF = np.sqrt(np.sum(np.square(x_wiener_ekf_estimate-x_wiener))/x_wiener.shape[0])
print('The rmse for UKF = {} and EKF={}'.format(RMSE_UKF,RMSE_EKF))
````

````python
f, ax = plt.subplots(1,2,figsize=(2*fig_size[0],fig_size[1]))
for i in range(2):
    ax[i].plot(t_wiener,x_wiener_ukf_estimate[:,i],linewidth=0.5, label='$\hat x_{}$ UKF'.format(i+1))
    ax[i].plot(t_wiener,x_wiener_ekf_estimate[:,i],linewidth=0.5, color='red', label='$\hat x_{}$ EKF'.format(i+1))
    ax[i].fill_between(t_wiener,x_wiener_ukf_estimate[:,i]-2*stdev_wiener_ukf[:,i],x_wiener_ukf_estimate[:,i]+2*stdev_wiener_ukf[:,i],label='conf. UKF \n $x_{}$'.format(i+1),color='b',alpha=0.2)
    ax[i].fill_between(t_wiener,x_wiener_ekf_estimate[:,i]-2*stdev_wiener_ekf[:,i],x_wiener_ekf_estimate[:,i]+2*stdev_wiener_ekf[:,i],label='conf. EKF \n $x_{}$'.format(i+1),color='r',alpha=0.2)
    ax[i].plot(t_wiener,x_wiener[:,i],'-k',linewidth=0.5, label='$x_{}$'.format(i+1))
    ax[i].set_xlabel('$t$')
    ax[i].set_ylabel('$x$')
    ax[i].legend()
````

Recall the Euler–Maruyama discretization of the 2D dynamic model of a robot platform in Exercise 8.3.

````python
robot_init = np.array([0.5,0.5,np.pi/4])
# robot_init = np.array([0.,0.,0.])
t_robot = np.linspace(0.,5.,501)
dt_robot = t_robot[1]-t_robot[0]
u_robot = np.zeros((t_robot.shape[0],2))

for i in range(t_robot.shape[0]):
    if 0<= t_robot[i] < 1:
        u_robot[i,0] = t_robot[i]
    elif 1<= t_robot[i] < 4:
        u_robot[i,0] = 1
    else:
        u_robot[i,0] = 5-t_robot[i]
        
    if 0<= t_robot[i] < 2:
        u_robot[i,1] = 0
    elif 2<= t_robot[i] < 3:
        u_robot[i,1] = np.pi
    else:
        u_robot[i,1] = 0
````

````python
Q_robot = np.diag([1e-2,1e-2,1e-2])
Q_robot_discrete = dt_robot*Q_robot
q_robot_discrete = np.random.multivariate_normal(np.zeros(3),Q_robot_discrete,(t_robot.shape[0]))
R_robot_discrete = 0.01*np.eye(2)
r_robot = np.random.multivariate_normal(np.zeros(2),R_robot_discrete,(t_robot.shape[0]))
````

````python
def _f_robot(x:np.ndarray,u:np.ndarray):
    return np.array([u[0]*np.cos(x[2]), u[0]*np.sin(x[2]), u[1]])

f_robot = np.vectorize(_f_robot,signature='(n),(m)->(n)')

def f_robot_discrete(x:np.ndarray,u:np.ndarray):
    return x + f_robot(x,u)*dt_robot

def df_dx_robot(x:np.ndarray,u:np.ndarray):
    jac = np.zeros((x.shape[0],x.shape[0]))
    jac[0,2] = -u[0]*np.sin(x[2])
    jac[1,2] = u[0]*np.cos(x[2])
    
    return jac

def df_dx_robot_discrete(x:np.ndarray,u:np.ndarray):
    return np.eye(x.shape[0])+df_dx_robot(x,u)*dt_robot

def _h_robot(x:np.ndarray):
    return np.array([x[0],x[1]])

h_robot = np.vectorize(_h_robot, signature='(n)->(2)') ##Vectorize the stuff to avoid doing looping



def dh_dx_robot(x:np.ndarray):
    return np.array([[1,0.,0.],[0.,1.,0.]])

def euler_maruyama_propagate(f:Callable[[np.ndarray,np.ndarray],np.ndarray], 
                              x_init:np.ndarray, w:np.ndarray, dt:float, u:np.ndarray=None):
    length = w.shape[0]
    if u is None:
        u = np.zeros((length,1))
    x_res = np.zeros((length,x_init.shape[0]))
    x_res[0] = x_init
    for i in range(x_res.shape[0]-1):
        x_res[i+1] = x_res[i] + f(x_res[i],u[i])*dt +  w[i]
    return x_res
````

````python
x_robot = euler_maruyama_propagate(f_robot, robot_init, q_robot_discrete, dt_robot, u_robot)
````

## Question 3 (Robot model with distance and bearing measurements)

g_fun = @(x) [sqrt((x(1)-sx)^2 + (x(2)-sx)^2); atan2(sy-x(2),sx-x(1))-x(3)];

## Assume that we have $(s_x , s_y ) = (0,0)$

````python
def _h_robot_2(x:np.ndarray):
    return np.array([np.sqrt((x[0]*x[0])+(x[1]*x[1]),np.arctan2(-x[1],x[0])-x[2]])

#h_robot_2 = np.vectorize(_h_robot, signature='(n)->(2)') ##Vectorize the stuff to avoid doing looping
h_robot_2 = np.vectorize(_h_robot_2, signature='(n)->(2)') ##Vectorize the stuff to avoid doing looping



def dh_dx_robot_2(x:np.ndarray):
    distance = np.sqrt(x[0]*x[0]+x[1]*x[1])
    return np.array([[x[0]/distance,x[1]/distance, 0.],
                     [-x[1]/distance,x[0]/distance, -1]])
````
