Import some standard libraries.

```python
## Solution by Muhammad Emzir, Aalto University, Oct 2020
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
sns.set()
```

Declare some variables
Depending on how you define the initial state and how big the measurement noises are, you might end up in a solution that is quite far from the truth.
Here, we use a fairly small noise variances. Change the noise variance $\sigma$

```python
#set some variables
N = 10 # number of observation set
sigma = 0.5 # we use the same standard deviation for all measurement
alpha = 10. #set alpha, and assume beta=alpha
x = np.array([.5,.5]) #set the true variable
x_0 = np.array([0.25,0.25]) #set the initial guess
iteration_end = 30 #how many iteration
x_history = np.zeros((iteration_end,2)) #the evolution of estimation results recorded here
x_history[0,:] = x_0 # you know this right?
R = sigma*sigma*np.eye(2) # measurement variance
R_inv = np.eye(2)/(sigma*sigma) #cache the inverse
E = sigma*np.eye(2) #we will use this below
```

Define the nonlinear measurmeent function $\mathbf g(\mathbf{x})$ and its Jacobian $\mathbf G(\mathbf{x})$

```python
def g(x):
    return alpha*np.sqrt(x)
```

```python
def G(x):
    return (alpha/2)*np.diag(1/np.sqrt(x))
```

Define the cost function $J$

```python
def J_cost(y,x):
    #This relies on the fact that our R is a block diagonal matrix, otherwise you need to use different notation
    g_ = g(x)
    e = (y-g_)@E
    return np.sum(e**2)
```

Simulate the measurements

```python
r = sigma*np.random.randn(N,2)
y = g(x)+ r
Grid_size=100
[x_grid,y_grid] = np.meshgrid(np.arange(Grid_size)/Grid_size,np.arange(Grid_size)/Grid_size)
J = np.zeros((Grid_size,Grid_size))
for i in range(x_grid.shape[0]):
    for j in range(x_grid.shape[1]):
        J[i,j] = J_cost(y,np.array([x_grid[i,j],y_grid[i,j]]))
```

We use a trick that in the case that $\mathbf{r}_i$ is independent each other we have
$$
(\bar{\mathbf{G}}^\top \bar{\mathbf{R}}^{-1} \bar{\mathbf{G}}    )^{-1}\bar{\mathbf{G}}^\top \bar{\mathbf{R}}^{-1}(\mathbf{y} - \bar{\mathbf{g}}(\mathbf{x}))
= (\sum_i\mathbf{G}^\top \mathbf R^{-1} \mathbf G    )^{-1}\sum_i\mathbf{G}^\top \mathbf{R}^{-1}(\mathbf{y}_i - \mathbf{g}(\mathbf{x}))
$$
Which equals to
$$
(\mathbf{G}^\top \mathbf R^{-1} \mathbf{G})^{-1}\dfrac{1}{N}\sum_i\mathbf{G}^\top \mathbf{R}^{-1}(\mathbf{y}_i - \mathbf{g}_i(\mathbf{x}))
$$

```python
for j in range(iteration_end-1):
    gj = g(x_history[j,:])
    Gj = G(x_history[j,:])
    x_history[j+1,:] = x_history[j,:] + np.linalg.solve((Gj.T@R_inv@Gj),Gj.T@R_inv@np.sum(y-gj,axis=0)/N)

```

Plot the estimation history

```python
fig = plt.figure(figsize=(15,15))
point_size=6
ax = fig.add_subplot(111)
ax.contour(x_grid,y_grid,J,levels=200)
ax.plot(x[0],x[1],c='b',marker="X")
ax.plot(x_history[:,0],x_history[:,1],c='k',marker="P",linewidth=0.5)
ax.set_xlim(0.25,0.75)
ax.set_ylim(0.25,0.75)
```
