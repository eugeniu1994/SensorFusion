```python
## Solution by Muhammad Emzir, Aalto University, Sept 2020
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
sns.set()
```

After standard python package import(s), we declare some of the following variables, which are the truth location of the drone $x$, measurement matrix $G$, and bias(es) $b$.

```python
x = np.array([0.5,0.5,0.5])#true position
L = 1.
I = np.eye(3)
G = np.vstack((I,-I))
b = np.array([0.,0.,0.,L,L,L])
```

Now we describe the measurements $y$, and the measurement errors $r$ and its covariance $R$.

```python
n_sensors = 6
stdev = 2e-1 #equal for all sensors
r  = stdev*np.random.randn(n_sensors)
R = stdev*stdev*np.eye(n_sensors)
y = G@x+b+r
```

Declare some variables for the prior $m$ and $P$. We deliberately precompute the inverse of $R$ and $P$

```python
#pick an m and P
m = np.zeros(3)
P = 1e-2*np.eye(3)# If you put P small, you better make sure that (x-m) is very close
P += P.T
R_inv = np.linalg.inv(R)
P_inv = np.linalg.inv(P)
```

Now compute the solution of the Weighted Least Square $\hat{x}_{WLS}$ and $\hat{x}_{RegWLS}$.

```python
y_tilde = y - b
x_WLS = np.linalg.solve(G.T@R_inv@G, G.T@R_inv@y_tilde)
x_reg_WLS = np.linalg.solve(G.T@R_inv@G+P_inv, G.T@R_inv@y_tilde+P_inv@m)
```

Compare the $L^2$ errors

```python
print("Error of the Least square is {}, Error of the regularized Least Square is {}". format(np.linalg.norm(x-x_WLS),np.linalg.norm(x-x_reg_WLS)))
```

The implementation of the sequential regularized weighted least square.

```python
x_est = m #put initially
x_history = np.zeros((7,3))
x_history[0,:] = x_est
P_est = P
for i in range(n_sensors):
    G_i = G[i,:]
    s = (np.inner(G_i,P_est@G_i)+R[i,i])
    K = P_est@G_i/s #explain why
    x_est = x_est + K*(y_tilde[i] - G_i@x_est)
    P_est = P_est - np.outer(K,K)*s
    x_history[i+1,:] = x_est

    
```

Check the update of the estimation. The estimation finally equals to the batch regularized weighted least square
