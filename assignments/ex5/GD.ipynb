```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_function(x):
    return (1.1-np.sin(x))**2

def d_cost_function(x): # function derivative
    return  -2.2*np.cos(x) + 2.*np.sin(x)*np.cos(x)
```

```python
xx = np.linspace(-10,10,1000)
cost = [cost_function(i) for i in xx]
grad = [d_cost_function(i) for i in xx]

plt.plot(xx,cost,label='J(x)')
plt.plot(xx,grad,label='gradient')
plt.legend()
plt.grid()
plt.show()
```

```python
max_iter = 110
precision = .001
```

```python
def gradient_descent(x0,lr=.1,precision=.1,max_iter=100):
    xi =[]
    loss_history=[]
    cost_history=[]
    
    x=x0
    
    xi.append(x)
    cost = cost_function(0)
    cost_history.append(cost)
    
    i=0
    
    while i<max_iter:
        gradient = d_cost_function(x)
        
        new_x = x-lr*gradient
        
        x = new_x
        xi.append(x)
        current_cost = cost_function(new_x)
        cost_history.append(current_cost)
        
        loss_history.append(abs(current_cost-cost))
        
        if abs(gradient) < precision:
            break
            
        i+=1
        
    print('Local minima x:{},  current_cost:{}'.format(x,current_cost))
    
    return loss_history, xi, i, cost_history
```

```python
loss_history, xi, i, cost_history = gradient_descent(0,.4,1e-4,1000)
```

```python
plt.plot(cost_history,'o-')
plt.grid()
plt.show()
```

```python
plt.plot(loss_history)
plt.grid()
plt.show()
```
